```python
import numpy as np
import pandas as pd
import plotly.express as ex 
import plotly.graph_objects as go
```

```python
path=r'C:\Users\Ali\Desktop\data.csv'
df=pd.DataFrame()

df=pd.read_csv(path)

df.head(5)
```

```python
x=df.drop('target', axis=1)
y=df['target']
```

```python
# supervised, SVM

from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# standardizing data and then splitting into train and test
x=StandardScaler().fit(x).transform(x)
x_train_svm,x_test_svm, y_train_svm, y_test_svm = train_test_split ( x ,y ,test_size=0.2,random_state=1) 


model_svm=svm.SVC(kernel='linear')


prediction=model_svm.fit(x_train_svm,y_train_svm)
prediction.predict(x_test_svm)
score_svm=prediction.score(x_test_svm, y_test_svm )

print('score: ', score_svm)
# model_svm.fit(x_train_svm, )
```

```python
count0=np.where(y_test_svm ==0)
count1=np.where(y_test_svm ==1)
np.array(count0).shape
np.array(count1).shape
```

```python
# supervised , linear regression , as expected , it was useless , it might however predict with some threshold values, I will try that later 



#now i understood that this was totally wrong for many reasons
from sklearn.linear_model import LinearRegression



model_linear=LinearRegression()

model_linear.fit(x_train_svm, y_train_svm)

score_linear= model_linear.score(x_test_svm,y_test_svm)

basic_prediction= pd.DataFrame(model_linear.predict(x_test_svm))

# 0.52 is the mean of predicted model score_linear 

basic_prediction_linear_class = np.where(basic_prediction[[0]] > 0.52 , 1, 0)

basic_prediction_linear_class = pd.DataFrame(basic_prediction_linear_class).astype('int64')

y_test_svm_df = pd.DataFrame(y_test_svm)


#apparently somewhere in the way I got an excessive dimension that i had to get rid of or mean() would not work

basic_prediction_linear_class = basic_prediction_linear_class.squeeze()
y_test_svm_df = y_test_svm_df.squeeze()

scre_linear_class= np.mean(np.abs(basic_prediction_linear_class - y_test_svm_df))

print('score: ', scre_linear_class)
```

```python

# unsupervised , KMeans , clustering  , well it did almost like svm 

from sklearn.cluster import KMeans

model_kmeans= KMeans(n_clusters=2, init='k-means++')
model_kmeans.fit(x_train_svm )

np.mean(np.abs(model_kmeans.predict(x_test_svm) - y_test_svm_df))
```
